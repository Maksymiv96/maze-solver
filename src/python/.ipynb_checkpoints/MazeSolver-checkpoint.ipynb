```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_name = 'simple'
img = cv.imread('../../imgs/{}.jpg'.format(img_name))
img = cv.resize()
b, g, r = cv.split(img)
rgb_img = cv.merge([r,g,b])
plt.imshow(rgb_img) 
plt.imshow(img)
plt.show()
```
